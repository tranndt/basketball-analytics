```python
HOME_PAGE = 'https://www.basketball-reference.com'
LEAGUES_PAGE = f'https://www.basketball-reference.com/leagues/'

# Scrape league seasons html
def request_all_league_seasons_html():
    html_text = request_html_text(LEAGUES_PAGE)
    html_text = clean_html_text(html_text)
    table = parse_html_table(html_text,extract_links='body')
    all_seasons_html_list = [href[1] for href in table.iloc[:,0].values]
    return all_seasons_html_list

def request_all_teams_per_league_season_html(league_season_html):
    season_standing = HOME_PAGE + league_season_html
    if 'standings' not in league_season_html:
        season_standing = season_standing.replace('.html','_standings.html')
    html_text = request_html_text(season_standing)
    html_text = clean_html_text(html_text)
    html_soup = parse_html_soup(html_text)
    ele = html_soup.find('table',id='team_vs_team')
    table = parse_html_table(ele.prettify(),extract_links='body')
    all_teams_per_league_season = [href[1] for href in table.iloc[:,1].values]
    return all_teams_per_league_season

all_league_seasons_html_dict = {}
for league_season_html in request_all_league_seasons_html():
    all_teams_per_league_season_html = request_all_teams_per_league_season_html(league_season_html)
    all_league_seasons_html_dict[league_season_html] = all_teams_per_league_season_html
    print(league_season_html)
    print(all_teams_per_league_season_html)
    clear_output(wait=True)
    time.sleep(3.1)

# Save dict to a txt file. Convert the dict to a string first
DST_DIR = './data-html'
save_file(f'{DST_DIR}/leagues_teams.txt',str(all_league_seasons_html_dict))
```
